```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def func(x, a, b):
    return 1 - np.exp(-(x / a) ** b)

# Generate a range of x values
x_values = np.linspace(0, 20, 500)

# Define some different a and b values
a_values = [1, 2, 3, 4, 5]
b_values = [1]

# Create a figure and axis to plot on
fig, ax = plt.subplots()

# Loop over all combinations of a and b values
for a in a_values:
    for b in b_values:
        # Calculate y values for this a and b
        y_values = func(x_values, a, b)

        # Plot these values with a label
        ax.plot(x_values, y_values, label=f'a={a}, b={b}')

# Add a legend and labels
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Function for different a and b')

# Display the plot
plt.show()
```
